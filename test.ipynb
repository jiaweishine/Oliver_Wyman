```python
def calculate_age(birth_date):
    if pd.isnull(birth_date):
        # 将缺失值替换为指定日期
        birth_date = '1900-01-01'
    # 将字符串日期转换为日期时间对象
    birth_date = datetime.datetime.strptime(birth_date, '%Y-%m-%d')
    
    # 获取当前日期
    current_date = datetime.datetime.now()
    
    # 计算年龄
    age = current_date.year - birth_date.year - ((current_date.month, current_date.day) < (birth_date.month, birth_date.day))
    
    return age
```

```python
def calculate_reject_stats(data):
    result_rows = []
    # 获取总记录数
    total_records = len(data)
    
    # 对于证件类型不为身份证的客户，予以拒绝；
    null_count = data['v_id_type_cd'].isnull().sum()
    reject_cnt=(data['v_id_type_cd']!=1).sum()
    reject_prob=reject_cnt/total_records
    result_rows.append({
            'name': 'v_id_type_cd',
            'null_cnt':null_count,
            'reject_cnt': reject_cnt,
            'reject_prob': reject_prob,
            'rule':'对于证件类型不为身份证的客户，予以拒绝'
        })
    
    # 借款人非外籍人士、非港澳台人士；

    null_count=data['V_NATIONALITY_COUNTRY_CD'].isnull().sum()
    reject_cnt=(data['V_NATIONALITY_COUNTRY_CD']!='CHN').sum()
    reject_prob=reject_cnt/total_records
    result_rows.append({
            'name': 'V_NATIONALITY_COUNTRY_CD',
            'null_cnt':null_count,
            'reject_cnt': reject_cnt,
            'reject_prob': reject_prob,
            'rule':'借款人非外籍人士、非港澳台人士'
        })
    

    # 证件类型是身份证,但借款人年龄小于18周岁或大于65周岁，予以拒绝；
    null_count=(data['v_id_type_cd'].isnull()|data['D_DATE_OF_BIRTH'].isnull()).sum()
    # 对 D_DATE_OF_BIRTH 列应用 calculate_age 函数
    data['age'] = data['D_DATE_OF_BIRTH'].apply(calculate_age)
    reject_cnt = ((data['v_id_type_cd'] == 1) & ((data['age'] < 18) | (data['age'] > 65))).sum()
    reject_prob=reject_cnt/total_records
    result_rows.append({
            'name': 'D_DATE_OF_BIRTH',
            'null_cnt':null_count,
            'reject_cnt': reject_cnt,
            'reject_prob': reject_prob,
            'rule':'证件类型是身份证,但借款人年龄小于18周岁或大于65周岁，予以拒绝'
        })

    # 创建包含所有结果的 DataFrame
    result_df = pd.DataFrame(result_rows)
    
    return result_df
```
